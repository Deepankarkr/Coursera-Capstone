## Steps in Data Processing
#### 1.Data Fetched from URL
#### 2.Data Converted to Dataframe
#### 3.Data Converted to Dataframe
#### 4.'Not Assigned' are filtered out from Borough
#### 5. For 'Not Assigned'Neighbourhood ,Borough is copied

```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
print("import Complete")
#-----------------------------
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(website_url,'lxml') 
My_table = soup.find('table',{'class':'wikitable sortable'})
data = []
print("Data Fetched from URL")
#-----------------------------
table_body = My_table.find('tbody')
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) 
A=['Postcode','Borough','Neighbourhood']
df = pd.DataFrame(data, columns=A)

print("Data Converted to Dataframe")
#-----------------------------

df2=df[df.Borough != 'Not assigned']

print("Data Converted to Dataframe")
#-----------------------------
df2=df[df.Borough != 'Not assigned']
print("'Not Assigned' are filtered out from Borough")
#-----------------------------

foo = lambda a: ",".join(a) 
df3=df2.groupby(['Postcode','Borough']).agg({'Neighbourhood':foo})
df3.reset_index(inplace=True)
for i in range(int(df3['Neighbourhood'].count())):
    if(list(df3[i:i+1]['Neighbourhood'])[0]=='Not assigned'):
      df3.at[i,'Neighbourhood']=df3.at[i,'Borough'] 
    
print(" For 'Not Assigned'Neighbourhood ,Borough is copied")

#-----------------------------
df3.shape
```
